# Individual model plots

```python
def accplothist(model, hist):
  # Summarize history for accuracy
  plt.plot(hist['accuracy'], color='blue', label='train')
  plt.plot(hist['val_accuracy'], color='orange', label='validation')
  plt.title((f'{model} Accuracy'))
  plt.ylabel('Accuracy')
  plt.xlabel('Epoch')
  plt.yticks(np.arange(80, 97, step=2))
  plt.xticks(np.arange(0, 16, step=1))
  plt.minorticks_on()
  plt.tick_params(axis='x', which='minor', bottom=False)
  plt.legend(['train', 'validation'], loc='lower right')
  plt.grid(color='gray', linestyle='dashed', linewidth=0.5, zorder=0)
  plt.savefig("/content/drive/MyDrive/Colab Notebooks/ViT-Base_Large-comparison_plot/vitl32-acc", dpi = 400, bbox_inches='tight')
  plt.show()

def lossplothist(model, hist):
  # Summarize history for loss
  plt.plot(hist['loss'], color='blue', label='train')
  plt.plot(hist['val_loss'], color='orange', label='validation')
  plt.title(f'{model} Categorical Cross-entropy Loss')
  plt.ylabel('Loss')
  plt.xlabel('Epoch')
  plt.yticks(np.arange(0.46, 0.76, step=0.02))
  plt.xticks(np.arange(0, 16, step=1))
  plt.minorticks_on()
  plt.tick_params(axis='x', which='minor', bottom=False)
  plt.legend(['train', 'validation'], loc='upper right')
  plt.grid(color='gray', linestyle='dashed', linewidth=0.5, zorder=0)
  plt.savefig("/content/drive/MyDrive/Colab Notebooks/ViT-Base_Large-comparison_plot/vitl32-loss", dpi = 400, bbox_inches='tight')
  plt.show()
```

**ViT-B/16**

```python
accplothist('ViT-B/16', histb16)
```

```python
lossplothist('ViT-B/16', histb16)
```

```python
lossplothist('ViT-B/32', histb32)
```

# Combined model plots

```python
def accplot_all_hist(hist1, hist2, hist3, hist4):
  # Summarize history for accuracy
  plt.plot(hist1['val_accuracy'], linestyle='solid', color='b', label='ViT-B/16')
  plt.plot(hist2['val_accuracy'], linestyle='dotted', color='orangered', label='ViT-B/32')
  plt.plot(hist3['val_accuracy'], linestyle='dashed', color='m', label='ViT-L/16')
  plt.plot(hist4['val_accuracy'], linestyle='dashdot', color='olivedrab', label='ViT-L/32')
  plt.title('ViT-Base and ViT-Large Validation Accuracy')
  plt.ylabel('Validation Accuracy')
  plt.xlabel('Epoch')
  plt.xticks(np.arange(0, 16, step=1))
  plt.legend(['ViT-B/16', 'ViT-B/32', 'ViT-L/16', 'ViT-L/32'], loc='lower right')
  plt.grid(color='gray', linestyle='dashed', linewidth=0.5, zorder=0)
  plt.savefig("/content/drive/MyDrive/Colab Notebooks/ViT-Base_Large-comparison_plot/allmodels-acc", dpi = 400, bbox_inches='tight')
  plt.show()

def  lossplot_all_hist(hist1, hist2, hist3, hist4):
  # Summarize history for loss
  plt.plot(hist1['val_loss'], linestyle='solid', color='b', label='ViT-B/16')
  plt.plot(hist2['val_loss'], linestyle='dotted', color='orangered', label='ViT-B/32')
  plt.plot(hist3['val_loss'], linestyle='dashed', color='m', label='ViT-L/16')
  plt.plot(hist4['val_loss'], linestyle='dashdot', color='olivedrab', label='ViT-L/32')
  plt.title('ViT-Base and ViT-Large Categorical Cross-entropy Loss')
  plt.ylabel('Validation Loss')
  plt.xlabel('Epoch')
  plt.xticks(np.arange(0, 16, step=1))
  plt.legend(['ViT-B/16', 'ViT-B/32', 'ViT-L/16', 'ViT-L/32'], loc='upper right')
  plt.grid(color='gray', linestyle='dashed', linewidth=0.5, zorder=0)
  plt.savefig("/content/drive/MyDrive/Colab Notebooks/ViT-Base_Large-comparison_plot/allmodels-loss", dpi = 400, bbox_inches='tight')
  plt.show()
```

```python
accplot_all_hist(histb16, histb32, histl16, histl32)
```

```python
lossplot_all_hist(histb16, histb32, histl16, histl32)
```
